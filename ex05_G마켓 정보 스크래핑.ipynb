### 실습! G마켓 Best 상품에서 Page에서 10ea의 상품 정보들을 스크래핑!
1. 크롬 브라우저 실행, url 요청(G마켓 Best 상품 Page)
2. 상품 클릭
3. 수집(상품명, 가격, 카테고리)
4. 뒤로가기
5. 2~4 단계 반복
6. DataFrame화

```python
driver = wb.Chrome()
driver.get('https://www.gmarket.co.kr/n/best')

cate_list = []
title_list = []
price_list = []

i = 0


while True :   
    try :
        img = driver.find_elements(By.CLASS_NAME, 'image__lazy.image__load')
        img[i].click()
        time.sleep(2)
        
        cate = driver.find_element(By.CSS_SELECTOR,
                           'body > div.location-navi > ul > li.on > a').text

        title = driver.find_element(By.CSS_SELECTOR,
                           '#itemcase_basic > div.box__item-title > h1').text

        price = driver.find_element(By.CSS_SELECTOR,
                           '#itemcase_basic > div.box__item-title > div.price > span:nth-child(2) > strong').text
        
        cate_list.append(cate)
        title_list.append(title)
        price_list.append(price)

        driver.back()
        time.sleep(2)
        
        i += 1

    except :
        break


import pandas as pd

dic = {'카테고리': cate_list, '상품': title_list, '가격' : price_list}

hansot = pd.DataFrame(dic)
hansot
```

```python
from selenium.webdriver.chrome.service import Service # 경로 반환
from webdriver_manager.chrome import ChromeDriverManager # 크롬 드라이버 관리자

import pandas as pd

from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup as bs

import time

from tqdm import tqdm as tq # 반복문 시각화하는 도구
```

```python
# 1. 크롬 브라우저 실행, url 요청(G마켓 Best 상품 Page)

s = Service(ChromeDriverManager().install())
driver = wb.Chrome(service=s)
driver.get("https://www.gmarket.co.kr/n/best")

# 5. 2~4 단계 반복
cate_list = []
title_list = []
price_list = []

for i in tq(range(50)) :
    # 2. 상품 클릭
    img = driver.find_elements(By.CLASS_NAME, "image__lazy.image__load")
    img[i].click()
    
    time.sleep(1) # 요소가 load될 때까지 로직 멈추기

    # 3. 수집(상품명, 가격, 카테고리) 및 추출
    cate = driver.find_element(By.CSS_SELECTOR, 
                               "div.location-navi > ul > li.on > a").text
    title = driver.find_element(By.CSS_SELECTOR, 
                               "#itemcase_basic > div.box__item-title > h1").text
    price = driver.find_element(By.CSS_SELECTOR, 
                               "#itemcase_basic > div.box__item-title > div.price > span:nth-child(2) > strong").text
    
    # 추출 후 list에 하나씩 쌓기!
    cate_list.append(cate)
    title_list.append(title)
    price_list.append(price)

    # 4. 뒤로가기
    driver.back()
    time.sleep(1) # 요소가 load될 때까지 로직 멈추기

# 자원 반납
driver.quit()
```

```python
# 6. DataFrame화

dic = {"카테고리" : cate_list,
       "상품명" : title_list,
       "가격" : price_list}

df = pd.DataFrame(dic)
df
```

```python
# csv 형식으로 저장!

# time.strftime("Y%m%d_%H%M%S")
# 현재 시점을 기준으로 연도, 월, 일, 시, 분, 초 문자열형으로 반환!
df.to_csv(f"{time.strftime('Y%m%d_%H%M%S')} G마켓 Bestesell Top 10.csv",
          encoding = 'euc-kr')
```
